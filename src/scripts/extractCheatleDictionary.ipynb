Open the source dictionary, remove all words less than 3 letters in length, and all words with non-alphabetic characters

```python
import pandas as pd

# Load JSON from source file
input_file_path = '../../static/dictionary.txt'
df = pd.read_csv(
    input_file_path, 
    sep=r"\s+", 
    header=None, 
    names=['word'],
    keep_default_na=False,
)

# Filter out non-numeric words, words shorter than 3 letters, and words longer than 16 letters (unless containing the 'QU' tile)
df_filtered = df[
    (
        (df['word'].str.len().between(3, 16)) &
        (df['word'].str.isalpha()) &
        (~df['word'].str.contains(r"q(?!u)", case=False, regex=True))
    )
    | (
        df['word'].str.contains("qu", case=False) &
        (df['word'].str.len() == 17)
    )
]
```

```python
# Save as ts file
output_file_path = '../data/dictionaryNew.ts'
words = df["word"].str.upper().tolist()

with open(output_file_path, "w") as f:
    f.write("export const VALID_WORD_DICTIONARY: string[] = [\n")
    f.write(",\n".join(f'  "{w}"' for w in words))
    f.write("\n];")
```
